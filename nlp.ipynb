```python
import nltk
```

```python
word_stem=nltk.stem.PorterStemmer()
print(word_stem.stem('dancing'))
print(word_stem.stem('writing'))
```

```python
wsreg=nltk.stem.RegexpStemmer('s')
wsreg.stem('dances')
```

```python
wsnow=nltk.stem.SnowballStemmer('spanish')
wsnow.stem('baile')
```

```python
nltk.download('punkt')
from nltk.tokenize import word_tokenize
t=word_tokenize("I'm not feeling well")
t
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
engstopwords=set(stopwords.words('english'))
words="No holiday for good friday".split(" ")
l=[word for word in words if word not in engstopwords]
l
```

```python
nltk.download('webtext')
import nltk
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import webtext
text = webtext.raw()
sent_tokenizer = PunktSentenceTokenizer(text)
sents_1 = sent_tokenizer.tokenize(text)
print(sents_1[0])
```

```python
nltk.download('wordnet')

from nltk.corpus import wordnet as wn
import nltk
syn = wn.synsets('drawing')[0]
syn.name()
```

```python
print(syn.definition())
print(syn.examples())
print(syn.hypernyms())
```
